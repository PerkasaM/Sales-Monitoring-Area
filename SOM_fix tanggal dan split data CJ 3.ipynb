```python
# ============================================================
# CELL 5 — FUNGSI HELPER
# ============================================================

STANDARD_COLS = [
    'CUSTOMER CODE1', 'CUSTOMER NAME', 'SDO', 'SKU NUMBER', 'SHORT CODE',
    'QTY', 'DESC', 'MASTERDATA.Description', 'MASTERDATA.Brand',
    'MASTERDATA.PG', 'MASTERDATA.SUB-PG', 'TOTAL ORDER (NETT * QTY)',
    'PO NUMBER', 'SO #', 'DR #', 'STATUS', 'REMARKS', 'AREA',
    'ORDER DATE', 'PROPOSE DATE TO CA', 'CYCLE', 'PAKET',
    'SPV NAME', 'ASM NAME', 'RSM NAME', 'CHANEL', 'REGION', 'PIC', 'SOURCE',
]


def clean_custcode(val, do_split):
    """Bersihkan customer code: buang .0 (float), split '/', strip whitespace."""
    if pd.isna(val): return None
    s = str(val).strip()
    if s.endswith('.0') and s[:-2].isdigit():
        s = s[:-2]
    if do_split and '/' in s:
        s = s.split('/')[0].strip()
    if s.endswith('.0') and s[:-2].isdigit():
        s = s[:-2]
    return s.strip()


def clean_numref(val):
    """Normalize angka referensi (DR#, SO#): 828236469.0 → '828236469', nan → ''"""
    if pd.isna(val): return ''
    s = str(val).strip()
    if s in ('nan', 'None', '-', '', 'NaN'): return ''
    if s.endswith('.0') and s[:-2].replace('-', '').isdigit():
        s = s[:-2]
    return s.strip()


def load_and_process(name, cfg):
    """Load satu template SOM dan ubah ke format standar."""
    print(f'\n🔄 Processing: {name}')

    dfs     = []
    files   = cfg['files']
    sheets  = cfg['sheet']  if isinstance(cfg['sheet'],  list) else [cfg['sheet']]  * len(files)
    headers = cfg['header'] if isinstance(cfg['header'], list) else [cfg['header']] * len(files)

    for i, fpath in enumerate(files):
        try:
            # usecols callable — hindari MemoryError (kolom nyasar ke kanan)
            # dan ValueError (header tipe campuran seperti sheet INDERECT)
            # Baca dengan data_only=True via openpyxl agar hasil formula (VLOOKUP) terbaca
            _ext = str(fpath).lower()
            if _ext.endswith('.xlsx') or _ext.endswith('.xlsm'):
                import openpyxl as _opxl
                _wb  = _opxl.load_workbook(fpath, data_only=True, read_only=True)
                _ws  = _wb[sheets[i]]
                _rows = list(_ws.iter_rows(values_only=True))
                _hdr  = headers[i]
                _cols = [str(c).strip() if c is not None else '' for c in _rows[_hdr]]
                _data = _rows[_hdr+1:]
                df    = pd.DataFrame(_data, columns=_cols)
                # Deduplikasi kolom — ambil kolom pertama jika ada nama sama
                seen = set()
                keep = []
                for c in df.columns:
                    if c and not c.startswith('Unnamed') and c != 'nan' and c not in seen:
                        keep.append(c)
                        seen.add(c)
                df = df[keep]
                _wb.close()
            else:
                df = pd.read_excel(
                    fpath,
                    sheet_name=sheets[i],
                    header=headers[i],
                    usecols=lambda c: str(c).strip() not in ('', 'nan')
                                      and not str(c).startswith('Unnamed'),
                )
            print(f'   ✅ {fpath.split(chr(92))[-1]} → {len(df)} baris, {len(df.columns)} kolom')

            # Normalize nama kolom
            df.columns = df.columns.astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)
            if cfg.get('strip_col_names'):
                df.columns = df.columns.str.upper()
            df = df.dropna(axis=1, how='all')
            df = df.loc[:, ~df.columns.str.contains('^Unnamed', na=False)]

            # Buang kolom duplikat (penyebab InvalidIndexError saat concat)
            df = df.loc[:, ~df.columns.duplicated()]
            dfs.append(df)
        except FileNotFoundError:
            print(f'   ⚠️  FILE TIDAK DITEMUKAN: {fpath}')
            print(f'       → Template {name} dilewati.')
            return None

    # Pastikan tiap df tidak punya kolom duplikat sebelum concat
    clean_dfs = []
    for _df in dfs:
        _seen = set()
        _keep = []
        for _c in _df.columns:
            _cs = str(_c).strip()
            if _cs and _cs != 'nan' and not _cs.startswith('Unnamed') and _cs not in _seen:
                _keep.append(_c)
                _seen.add(_cs)
        clean_dfs.append(_df[_keep])
    df = pd.concat(clean_dfs, ignore_index=True) if len(clean_dfs) > 1 else clean_dfs[0]

    # Fix format tanggal serial Excel
    # Konversi kolom numerik agar groupby sum tidak error
    for _num_col in ['QTY', 'TOTAL ORDER (NETT * QTY)']:
        if _num_col in df.columns:
            df[_num_col] = pd.to_numeric(
                df[_num_col].astype(str)
                  .str.replace(r'[^\d.\-]', '', regex=True)
                  .replace('', '0'),
                errors='coerce'
            ).fillna(0)

    if cfg.get('fix_order_date') and 'ORDER DATE' in df.columns:
        col = df['ORDER DATE']
        # Konversi numerik — clamp ke range valid serial date Excel (1 s/d 60000)
        # agar tidak OutOfBoundsDatetime dari nilai yang bukan tanggal
        num_col = pd.to_numeric(col, errors='coerce')
        valid_serial = num_col.where((num_col >= 1) & (num_col <= 60000))
        date_from_serial = pd.to_datetime(valid_serial, unit='D', origin='1899-12-30', errors='coerce')
        date_from_slash  = pd.to_datetime(col, format='%d/%m/%Y', errors='coerce')
        date_from_dash   = pd.to_datetime(col, format='%d-%m-%Y', errors='coerce')
        df['ORDER DATE'] = (
            date_from_serial
            .fillna(date_from_slash)
            .fillna(date_from_dash)
            .dt.strftime('%d/%m/%Y')
        )

    # Buang baris yang terlalu banyak kosong
    df = df.dropna(thresh=int(df.shape[1] * 0.5))

    # Filter cycle
    cycle_col = cfg.get('cycle_col', 'CYCLE')
    cycle_val = cfg.get('cycle_val', DEFAULT_CYCLE)
    if cycle_col in df.columns:
        df = df[df[cycle_col] == cycle_val]
        print(f'   📌 Setelah filter cycle \'{cycle_val}\': {len(df)} baris')

    # Filter SDO (opsional) — hanya ambil SDO tertentu dari file ini
    sdo_filter = cfg.get('sdo_filter')
    if sdo_filter and 'SDO' in df.columns:
        sdo_upper = [s.upper().strip() for s in sdo_filter]
        df = df[df['SDO'].astype(str).str.upper().str.strip().isin(sdo_upper)]
        print(f'   🎯 Setelah filter SDO: {len(df)} baris')

    df = df.reset_index(drop=True)

    # Mapping kolom ke format standar
    rekapsom = pd.DataFrame(index=df.index, columns=STANDARD_COLS)
    for src_col, dst_col in cfg['col_map'].items():
        if src_col in df.columns and rekapsom[dst_col].isna().all():
            rekapsom[dst_col] = df[src_col].values

    # Static columns
    rekapsom['CHANEL']  = cfg['chanel']
    rekapsom['REMARKS'] = rekapsom['STATUS']
    rekapsom['SOURCE']  = name
    for col, val in cfg.get('static_cols', {}).items():
        if val is not None:
            rekapsom[col] = val
        elif col in ('REGION', 'PIC') and 'AREA' in rekapsom.columns:
            rekapsom[col] = rekapsom['AREA']

    # Bersihkan Customer Code
    do_split = cfg.get('split_custcode', False)
    if 'CUSTOMER CODE1' in rekapsom.columns:
        rekapsom['CUSTOMER CODE1'] = rekapsom['CUSTOMER CODE1'].apply(
            lambda v: clean_custcode(v, do_split)
        )

    # SDO uppercase
    if cfg.get('sdo_upper') and 'SDO' in rekapsom.columns:
        rekapsom['SDO'] = rekapsom['SDO'].astype(str).str.upper()

    # SDO rename (exact match)
    for old, new in cfg.get('sdo_rename', {}).items():
        rekapsom.loc[rekapsom['SDO'] == old.upper(), 'SDO'] = new

    # SDO clean (exact atau contains)
    if cfg.get('sdo_clean'):
        mode = cfg.get('sdo_clean_mode', 'exact')
        for keyword, replacement in cfg['sdo_clean'].items():
            if mode == 'contains':
                mask = rekapsom['SDO'].str.contains(keyword.upper(), na=False)
            else:
                mask = rekapsom['SDO'] == keyword.upper()
            rekapsom.loc[mask, 'SDO'] = replacement

    # Standarisasi tipe data
    rekapsom['CUSTOMER CODE1'] = rekapsom['CUSTOMER CODE1'].astype(str).str.strip()
    rekapsom['SKU NUMBER']     = rekapsom['SKU NUMBER'].astype(str).str.strip()

    return rekapsom


def enrich_masterdata(df):
    """Tambahkan SPV/ASM/RSM, SubPG, Group, dan Numcode."""

    # SPV-RSM
    df = df.merge(spvrsm[['SDO', 'spv', 'asm', 'rsm']], on='SDO', how='left')
    df['SPV NAME'] = df['SPV NAME'].fillna(df.get('spv'))
    df['ASM NAME'] = df['ASM NAME'].fillna(df.get('asm'))
    df['RSM NAME'] = df['RSM NAME'].fillna(df.get('rsm'))
    df = df.drop(columns=[c for c in ['spv', 'asm', 'rsm'] if c in df.columns])

    # SubPG
    if 'SHORT CODE' in df.columns and 'SUBPG2' in subpg1.columns:
        df = df.merge(subpg1[['SHORT CODE', 'SUBPG2']], on='SHORT CODE', how='left')
        df = df.rename(columns={'SUBPG2': 'SUBPG1'})

    # Group
    if 'CUSTOMER CODE1' in df.columns:
        grp_cols = ['CUSTOMER CODE1', 'GROUP']
        if 'GROUP_REGION' in group.columns:
            grp_cols.append('GROUP_REGION')
        df = df.merge(group[grp_cols], on='CUSTOMER CODE1', how='left')

        # Isi AREA dan REGION dari GROUP_REGION untuk MODERN TRADE dan PROJECTS
        if 'GROUP_REGION' in df.columns:
            mask_mt_proj = df['CHANEL'].isin(['MODERN TRADE', 'PROJECTS'])
            has_region   = mask_mt_proj & df['GROUP_REGION'].notna() & (df['GROUP_REGION'] != '')
            df.loc[has_region, 'AREA']   = df.loc[has_region, 'GROUP_REGION']
            df.loc[has_region, 'REGION'] = df.loc[has_region, 'GROUP_REGION']
            df.loc[has_region, 'PIC']    = df.loc[has_region, 'GROUP_REGION']
            df = df.drop(columns=['GROUP_REGION'])
            n_updated = has_region.sum()
            print(f'   ✅ AREA/REGION MT & PROJECTS diupdate dari Group master: {n_updated:,} baris')

    # Numcode
    if 'SKU NUMBER' in df.columns:
        df = df.merge(
            numcode[['SKU NUMBER', 'Description', 'Brand', 'PG', 'SUB-PG']],
            on='SKU NUMBER', how='left',
        )
        df['MASTERDATA.Description'] = df['MASTERDATA.Description'].fillna(df.get('Description'))
        df['MASTERDATA.Brand']       = df['MASTERDATA.Brand'].fillna(df.get('Brand'))
        df['MASTERDATA.PG']          = df['MASTERDATA.PG'].fillna(df.get('PG'))
        df['MASTERDATA.SUB-PG']      = df['MASTERDATA.SUB-PG'].fillna(df.get('SUB-PG'))
        df = df.drop(columns=[c for c in ['Description', 'Brand', 'PG', 'SUB-PG'] if c in df.columns])

    if 'DESC' in df.columns and 'MASTERDATA.Description' in df.columns:
        df['DESC'] = df['DESC'].fillna(df['MASTERDATA.Description'])

    return df


print('✅ Fungsi helper berhasil didefinisikan')
```

```python
# ============================================================
# CELL 6 — 🚀 PROSES SEMUA TEMPLATE
# ============================================================

all_rekap = []
skipped   = []

for name, cfg in FILE_CONFIG.items():
    result = load_and_process(name, cfg)
    if result is not None and len(result) > 0:
        all_rekap.append(result)
    else:
        skipped.append(name)

print(f'\n{"="*50}')
print(f'✅ Berhasil diproses : {len(all_rekap)} template')
if skipped:
    print(f'⚠️  Dilewati         : {", ".join(skipped)}')
```

```python
# ============================================================
# CELL 7 — GABUNGKAN + ENRICH MASTER DATA
# ============================================================

print('🔗 Menggabungkan semua data...')
gabungan = pd.concat(all_rekap, ignore_index=True)
print(f'   Total baris gabungan : {len(gabungan):,}')
print(f'   Per source           : {gabungan["SOURCE"].value_counts().to_dict()}')

print('\n🔗 Menggabungkan dengan Master Data...')
gabungan = enrich_masterdata(gabungan)
print(f'✅ Selesai — {len(gabungan):,} baris, {len(gabungan.columns)} kolom')
```

```python
# CELL 8A — CLEANING STATUS

def cleaning_status(s):
    if pd.isna(s): return None
    s = str(s).upper().strip()
    if any(k in s for k in ['CANCEL', 'CLOSE', 'REJECT']): return 'CANCEL'
    elif 'OOS' in s or 'OUT OF STOCK' in s or 'OUT OFF STOCK' in s:                return 'OUT OF STOCK'
    elif s == 'DC' or 'OOS DC' in s:                       return 'DISCOUNTINUE'
    elif 'PENDING' in s:                                    return 'PENDING'
    elif 'ON PROSES BY SSO' in s or 'PROCC' in s or 'PROC' in s: return 'ON PROCESS BY SSO'
    elif 'RIVIEW PAK' in s:                                 return 'CHECK BY CREDIT'
    elif 'OVER LIMIT' in s:                                 return 'OVERLIMIT'
    else:                                                   return s

gabungan['STATUS']  = gabungan['STATUS'].astype(str).apply(cleaning_status)
gabungan['REMARKS'] = gabungan['STATUS']
print('✅ STATUS:', gabungan['STATUS'].value_counts().to_dict())
```

```python
# CELL 8B — CLEANING SDO

def cleaning_sdo_global(s):
    if pd.isna(s): return None
    s = str(s).upper().strip()
    if 'ANDY WIJAYA'        in s: return 'ANDY'
    elif 'IMAM SHOVI'       in s: return 'IMAM SHOVII'
    elif 'OKTO'             in s: return 'OKTO RISMAN'
    elif 'ANDI MANGGALA' in s or 'ANDI ARDIYANSYAH' in s: return 'ANDI MANGGALA PUTRA'
    elif 'JINO'             in s: return 'FREEKY JINO'
    elif 'MUHAMMAD LANANG GALIH' in s :return 'MUHAMMAD LANANG GALIH GUMILANG'
    elif 'ABDUL JABBAR'         in s:return 'ABDUL JABAR'
    else:                         return s

gabungan['SDO'] = gabungan['SDO'].apply(cleaning_sdo_global)
print('✅ SDO cleaning selesai')
```

```python
# CELL 8C — CLEANING REGION & AREA

# Mapping berdasarkan ASM NAME
asm_map = {'JOHANNES ANDRIYANTO': 'EAST JAVA 1', 'YANUAR AFANDI': 'EAST JAVA 2', 'IWAN SUSANTO': 'CENTRAL JAVA 2'}
mask = gabungan['ASM NAME'].str.upper().str.strip().isin(asm_map)
mapped = gabungan.loc[mask, 'ASM NAME'].str.upper().str.strip().map(asm_map)
for col in ['REGION', 'AREA', 'PIC']:
    gabungan.loc[mask, col] = mapped.values
gabungan.loc[gabungan['ASM NAME'].str.upper().str.strip() == 'IWAN SUSANTO', 'RSM NAME'] = 'BLANK'

# Mapping berdasarkan SDO
sdo_map = {
    'ANWAR KAMIL SURYONEGORO': 'BALINUSTRA', 'ABDUL JABAR': 'SULAWESI',
    'MULYADI': 'SULAWESI', 'ANDI MANGGALA PUTRA': 'SULAWESI',
    'RISKI WAHYUDI': 'SULAWESI', 'JEKY TIRTA': 'SULAWESI',
    'MOHAMMAD HENDRA FARIZAL':'BALINUSTRA'
}
mask = gabungan['SDO'].str.upper().str.strip().isin(sdo_map)
mapped = gabungan.loc[mask, 'SDO'].str.upper().str.strip().map(sdo_map)
for col in ['REGION', 'AREA', 'PIC']:
    gabungan.loc[mask, col] = mapped.values

# Mapping berdasarkan RSM NAME
rsm_map = {'IMAM SHOVII': 'CENTRAL JAVA 3', 'ANDI HARIS': 'CENTRAL JAVA 1'}
mask = gabungan['RSM NAME'].str.upper().str.strip().isin(rsm_map)
mapped = gabungan.loc[mask, 'RSM NAME'].str.upper().str.strip().map(rsm_map)
for col in ['REGION', 'PIC']:
    gabungan.loc[mask, col] = mapped.values

# Koreksi khusus
gabungan.loc[gabungan['SDO'] == 'MDI',              ['CHANEL','PIC']]      = 'DIRECT MDI'
gabungan.loc[gabungan['SDO'] == 'MDI',              ['AREA','REGION']]     = 'JABODETABEK'
gabungan.loc[gabungan['SDO'] == 'ANDIKA RAMADHAN',  ['AREA','PIC']]        = 'SUMATERA DIRECT'
gabungan.loc[gabungan['SDO'] == 'ANDIKA RAMADHAN',  'REGION']              = 'JAMBI'
gabungan.loc[gabungan['SDO'] == 'ADEK AGUSTIAN',    ['AREA','PIC']]        = 'SUMATERA DIRECT'
gabungan.loc[gabungan['SDO'] == 'ADEK AGUSTIAN',    'REGION']              = 'JAMBI'
gabungan.loc[gabungan['SDO'] == 'HAMDANI', ['AREA', 'PIC']]                = 'SUMATERA DIRECT'
gabungan.loc[gabungan['SDO'] == 'HAMDANI', 'REGION']                       = 'JAMBI' 
gabungan.loc[gabungan['RSM NAME'] == 'IMAM SHOVII', 'CHANEL']              = 'DIRECT'
gabungan.loc[gabungan['PIC'] == 'EAST INDONESIA',   'PIC']                 = 'BALINUSTRA'
gabungan.loc[gabungan['ASM NAME'] == 'FREEKY JINO', 'REGION']              = 'PANTURA'
for cust, val in [
    ('TOTAL OPTIMA PERKAKAS',             'KALIMANTAN'),
    ('PT INDOSUPPLY UNIVERSAL PERKASA',   'SUMATERA'),
    ('CAHAYA YURFA SEJAHTERA',            'SUMATERA'),
    ('ASIA ELEKTRIKAL CENTER',            'SUMATERA'),
]:
    gabungan.loc[gabungan['CUSTOMER NAME'] == cust, ['REGION','PIC','AREA']] = val

# Normalisasi AREA
def cleaning_area(a):
    if pd.isna(a): return None
    a = str(a).upper().strip()
    if 'JABO'           in a: return 'JABODETABEK'
    elif 'EAST JAVA'    in a: return 'EAST JAVA'
    elif 'CENTRAL'      in a: return 'CENTRAL JAVA'
    elif 'EAST INDONESIA' in a: return 'BALINUSTRA'
    elif 'SULAWESI'     in a: return 'SULAWESI DIRECT'
    else:                     return a

def cleaning_region(r):
    if pd.isna(r): return None
    r = str(r).upper().strip()
    if 'JABO'             in r: return 'JABODETABEK'
    elif 'EAST INDONESIA' in r: return 'BALINUSTRA'
    elif 'SULAWESI'       in r: return 'SULAWESI'
    else:                       return r

gabungan['AREA']   = gabungan['AREA'].apply(cleaning_area)
gabungan['REGION'] = gabungan['REGION'].apply(cleaning_region)
print('✅ AREA:',   gabungan['AREA'].value_counts().head(8).to_dict())
print('✅ REGION:', gabungan['REGION'].value_counts().head(8).to_dict())
```

```python
# CELL 8D — CLEANING PAKET

def cleaning_paket(p):
    if pd.isna(p): return p
    s = str(p).upper().strip()
    if '&'           in s:                                         return s
    elif 'GEBYAR'    in s:                                         return 'PROGRAM GEBYAR'
    elif 'END YEAR'  in s or 'END OF YEAR' in s or 'SUPERDEAL' in s: return 'PROGRAM END YEAR DEALS'
    elif 'CAHAYA'    in s:                                         return 'PROGRAM CAHAYA'
    elif 'MAJU'      in s:                                         return 'PROGRAM MAJU'
    elif 'HEBAT'     in s:                                         return 'PROGRAM HEBAT'
    elif 'RAMADHAN'  in s:                                         return 'PROGRAM RAMADHAN'
    else:                                                          return s

gabungan['PAKET'] = gabungan['PAKET'].apply(cleaning_paket)
print('✅ PAKET:', gabungan['PAKET'].value_counts().head(8).to_dict())
```

```python
# CELL 8E — CLEANING SUBPG1 + NORMALISASI CYCLE

main_kategori = ['BULB', 'CAPSULE', 'DOWNLIGHT', 'FLOODLIGHT', 'SWM']

def cleaning_subpg1(s):
    if pd.isna(s): return None
    s = str(s).upper().strip()
    if 'MINIMALIST' in s: return 'SWM'
    elif 'DOWNLIGHT' in s: return 'DOWNLIGHT'
    elif 'CAPSULE'   in s: return 'CAPSULE'
    elif 'BULB'      in s: return 'BULB'
    else:                  return s

if 'SUBPG1' in gabungan.columns:
    gabungan['SUBPG1'] = gabungan['SUBPG1'].fillna(gabungan.get('MASTERDATA.SUB-PG'))
    gabungan['SUBPG1'] = gabungan['SUBPG1'].apply(cleaning_subpg1)
    gabungan['SUBPG1'] = gabungan['SUBPG1'].str.upper().where(
        lambda x: x.isin(main_kategori), 'OTHERS'
    )
    print('✅ SUBPG1:', gabungan['SUBPG1'].value_counts().to_dict())
else:
    print('⚠️  Kolom SUBPG1 tidak ditemukan, dilewati.')

gabungan['CYCLE'] = 'C052026'
print('✅ CYCLE dinormalisasi → C052026')
```

```python
# ============================================================
# CELL 10 — 💾 SIMPAN KE EXCEL
# ============================================================

FINAL_COLS = [
    'SOURCE',
    'CUSTOMER CODE1', 'CUSTOMER NAME', 'SDO',
    'SKU NUMBER', 'SHORT CODE', 'QTY',
    'DESC', 'MASTERDATA.Description', 'MASTERDATA.Brand', 'MASTERDATA.PG', 'MASTERDATA.SUB-PG',
    'TOTAL ORDER (NETT * QTY)',
    'PO NUMBER', 'SO #', 'DR #', 'STATUS', 'REMARKS', 'AREA',
    'ORDER DATE', 'PROPOSE DATE TO CA', 'CYCLE', 'PAKET',
    'SPV NAME', 'ASM NAME', 'RSM NAME', 'CHANEL', 'REGION', 'PIC', 'SUBPG1', 'GROUP',
]

final_cols_exist = [c for c in FINAL_COLS if c in gabungan.columns]
output_df = gabungan[final_cols_exist]

# ── PASTIKAN QTY & TOTAL ORDER NUMERIC (fix groupby sum) ────────────────────
for _col in ['QTY', 'TOTAL ORDER (NETT * QTY)']:
    if _col in gabungan.columns:
        gabungan[_col] = pd.to_numeric(
            gabungan[_col].astype(str).str.replace(r'[^\d.\-]', '', regex=True).replace('', '0'),
            errors='coerce'
        ).fillna(0)

# ── CEK DOUBLE ENTRY ─────────────────────────────────────────────────────────
# Key: CUSTOMER CODE + ORDER DATE + SHORT CODE + QTY + TOTAL ORDER + REF (DR#→SO#→PO)
df_chk = output_df.copy()

# PERBAIKAN 1: normalize TOTAL ORDER ke integer string SEBELUM buat key
df_chk['_TOTAL_NORM'] = pd.to_numeric(
    df_chk['TOTAL ORDER (NETT * QTY)'].astype(str)
    .str.replace(r'[^\d.\-]', '', regex=True).replace('', '0'),
    errors='coerce'
).fillna(0).round(0).astype(int).astype(str)

# PERBAIKAN 2: normalize DR # dan SO # — buang .0 dari float
df_chk['_DR_NORM'] = df_chk['DR #'].apply(clean_numref)
df_chk['_SO_NORM'] = df_chk['SO #'].apply(clean_numref)

# REF: DR# → fallback SO# → fallback PO NUMBER
df_chk['_REF'] = df_chk['_DR_NORM']
mask_no_dr = df_chk['_REF'] == ''
df_chk.loc[mask_no_dr, '_REF'] = df_chk.loc[mask_no_dr, '_SO_NORM']
mask_no_so = df_chk['_REF'] == ''
df_chk.loc[mask_no_so, '_REF'] = df_chk.loc[mask_no_so, 'PO NUMBER'].astype(str).str.strip()

# PERBAIKAN 3: key lebih kuat — tambah CUSTOMER CODE + QTY
df_chk['_KEY'] = (
    df_chk['CUSTOMER CODE1'].astype(str).str.strip() + '|' +
    df_chk['ORDER DATE'].astype(str).str.strip() + '|' +
    df_chk['SHORT CODE'].astype(str).str.strip() + '|' +
    df_chk['QTY'].astype(str).str.strip() + '|' +
    df_chk['_TOTAL_NORM'] + '|' +
    df_chk['_REF']
)

# Tandai duplikat — keep='first' → baris pertama dianggap asli, sisanya = DOUBLE
df_chk['_IS_DOUBLE'] = df_chk.duplicated(subset=['_KEY'], keep='first')

# Sheet DOUBLE ENTRY: hanya baris duplikat (ke-2 dst)
drop_cols = ['_TOTAL_NORM', '_DR_NORM', '_SO_NORM', '_REF', '_KEY', '_IS_DOUBLE']
double_df = df_chk[df_chk['_IS_DOUBLE']].drop(columns=drop_cols)
n_double  = len(double_df)

with pd.ExcelWriter(OUTPUT_FILE, engine='openpyxl') as writer:
    output_df.to_excel(writer, sheet_name='SOM TERPADU', index=False)
    summary = gabungan.groupby(['SOURCE','CHANEL']).agg(
        TOTAL_BARIS=('CUSTOMER CODE1', 'count'),
        TOTAL_QTY=('QTY', 'sum'),
    ).reset_index()
    summary.to_excel(writer, sheet_name='RINGKASAN', index=False)
    if n_double > 0:
        double_df.to_excel(writer, sheet_name='DOUBLE ENTRY', index=False)

print(f'✅ Tersimpan: {OUTPUT_FILE}')
print(f'   Sheet SOM TERPADU : {len(output_df):,} baris')
if n_double > 0:
    print(f'   ⚠️  Sheet DOUBLE ENTRY : {n_double:,} transaksi double ditemukan!')
else:
    print(f'   ✅ Tidak ada transaksi double')
print(f'🎉 Selesai!')
```
